## Merging the datasets movies and credits and named the new dataset as movies

```python
movies = movies.merge(credits,on='title')
```

## Choosing columns which are helping and storing them into the movies dataset

```python
movies = movies[['movie_id','title','genres','keywords','overview','cast','crew']]
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies = movies.dropna()
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies['keywords'].iloc[0]
```

```python
import ast
```

## Extracting only names of genres from genres column 

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

## Extracting only names of keywords from keywords column 

```python
def convert(obj):
    return [i['name'] for i in ast.literal_eval(obj)]
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

## Extracting only names of cast from cast column 

```python
def convert3(obj):
    return [i['name'] for i in ast.literal_eval(obj)[:3]]

# Second way of defining function
# def convert3(obj):
#    L=[]
#    counter=0
#    for i in ast.literal_eval(obj):
#        if counter != 3:
#            L.append(i['name'])
#        else:
#           break
#    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

## Extracting only names of director from crew column 

```python
movies['crew'].iloc[0]
```

```python
def director(obj):
    L =[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head(1)
```

## Removing spaces from in between of two words like sam worthington will be samworthington

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1

# Using Lambda 

# movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['cast']
```

## Spliting the string of overview column into list, so that I can add it with other list columns(keywords, cast, genres, crew)

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview']
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies['tags']
```

## Dropping the old columns beacuse I created new column named "Tags" which contains values of old columns and storing it into new table named "new_movies"

```python
new_movies = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
new_movies.head(1)
```

## Converting the list into the String using Lambda 

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
new_movies.head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: x.lower())
```

```python
new_movies.head(1)
```

## Now, Converting tags into vector. For that CountVectorizer class of SKlearn will be used

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
similarity.shape
```

## Now, creating a function named recommend so that if one type a movie name, it will suggest 5 similar movies

```python
def recommend(movie):
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    distances = similarity[movie_index]

    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    for i in movie_list:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend('Batman')
```
